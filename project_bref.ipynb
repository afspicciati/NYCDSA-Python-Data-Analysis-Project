```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno
```

```python
war_batting = pd.read_csv('war_daily_bat.txt')
```

```python
war_batting.keys()
```

```python
missingno.bar(war_batting, figsize=(15,8), fontsize=12)
```

```python
plt.bar(war_batting[~war_batting['salary'].isnull()].groupby(by='year_ID').count()['player_ID'].index, war_batting[~war_batting['salary'].isnull()].groupby(by='year_ID').count()['player_ID'])
```

```python
plt.bar(war_batting[war_batting['salary'].isnull()].groupby(by='year_ID').count()['player_ID'].index, war_batting[war_batting['salary'].isnull()].groupby(by='year_ID').count()['player_ID'])
```

```python
plt.bar(war_batting[~war_batting['salary'].isnull()].groupby(by='year_ID').sum('G')['G'].index, war_batting[~war_batting['salary'].isnull()].groupby(by='year_ID').sum('G')['G'])
```

```python
plt.bar(war_batting[war_batting['salary'].isnull()].groupby(by='year_ID').sum('G')['G'].index[-21:], war_batting[war_batting['salary'].isnull()].groupby(by='year_ID').sum('G')['G'][-21:])
```
